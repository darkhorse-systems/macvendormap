```python
from urllib.parse import quote
import requests
from concurrent.futures import ThreadPoolExecutor
from IPython.display import HTML, display
import tabulate
import time
import random

oui_vendor_map = {}


def parse_text_from_nac(text):
    lines = []
    for line in text.split("\n"):
        line = line.strip()
        if line:
            parts = [e.strip() for e in line.split("\t") if e.strip()]
            lines.append(parts)
    return lines


def parse_addresses(mac_addresses):
    return [a.strip() for a in mac_addresses.split() if a.strip()]


def get_oui(mac_address):
    return "".join(mac_address.split(":")[:3])


def get_vendor_from_mac_address(mac_address):
    oui = get_oui(mac_address)
    if oui in oui_vendor_map:
        return oui_vendor_map[oui]
    url = f"https://api.macvendors.com/{quote(mac_address)}"
    vendor = requests.get(url).text
    oui_vendor_map[oui] = vendor
    # slow the requests to avoid spamming the free website
    time.sleep(1)
    return vendor


def get_vendor_from_mac_addresses(mac_addresses, workers=1):
    if type(mac_addresses) == str:
        mac_addresses = parse_addresses(mac_addresses)
    with ThreadPoolExecutor(max_workers=workers) as executor:
        result = executor.map(get_vendor_from_mac_address, mac_addresses)
    return [e for e in result]


def print_table(table):
    display(HTML(tabulate.tabulate(table, tablefmt='html')))
    
    
def main(text_from_nac=text_from_nac):
    table = parse_text_from_nac(text_from_nac)
    result_table = []
    for row in table:
        mac_address = row[0]
        vendor = get_vendor_from_mac_address(mac_address)
        new_row = [row[0], vendor] + row[1:]
        result_table.append(new_row)
    print_table(result_table)


def munge_mac(mac):
    parts = mac.split(":")
    new_mac = parts[:3] + [hex(random.randrange(0,256))[2:].upper() for e in range(3)]
    return ":".join(new_mac)
    
def munge_macs():
    for line in text_from_nac.split("\n"):
        if line:
            parts = [e.strip() for e in line.split("\t")]
            nm = munge_mac(parts[0])
            new_line = "\t".join([nm] + parts[1:])
            print(new_line)
    
    
# go:
main()
```
